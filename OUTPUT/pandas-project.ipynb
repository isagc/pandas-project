```python
import pandas as pd
```

```python
df = pd.read_csv('./INPUT/GSAF5.csv', encoding = "cp1252") 
```

```python
df.head()
```

```python
'''
Empiezo por comprobar si todos los datos de cada columna estan metidos en la base de datos 
en la clase que les corresponde
'''
df.dtypes
```

```python
'''
Una vez que he comprobado que todo estaba bien, voy a ver si puedo eliminar alguna columna de por si 
por contener muchos nulls. Primero, utilizo la funcion .isnull().sum() para que me encuentre cuantos nulos hay en
cada columna y que me los sume.
'''
null_cols = df.isnull().sum()
null_cols
```

```python
'''
Para ver con que criterio elimino columnas con muchos nulos, averiguo el numero de datos por columna y veo
si se aproxima mucho al numero de nulos, no me merece la pena continuar con esas columnas en mi analisis.
Al ver que las columnas Unnamed:22' & 'Unnamed:23' tan solo continen una y dos filas, respectivamente,
con valores que no sean nulos, las elimino.
'''
drop_cols = list(null_cols[null_cols > 5000].index)
df = df.drop(drop_cols, axis=1)
print(df.shape)
```

```python
df.head()
```

```python
'''
Aun asi me queda un df demasiado grande y me molesta bastante no ver el nombre de todas las columnas
por lo que elimine todas aquellas que yo categorice como "no relevantes" y son las siguientes:'pdf', 'href formula', 'href',
'Case Number.1','Case Number.2'(porque hay una tercera que se llama 'Case Number' y tiene los mismos datos y 'Name'
'''
df = df.drop(['pdf', 'href formula', 'href','Case Number.1','Case Number.2', 'Name'], axis = 1)
```

```python
df
```

```python
'''
Como tambien tenia una infinidad de filas, y me fije que habia datos de aÃ±os muy del pasado (como por ejemplo alguno de B.C)
y pense que eran poco fiables, me he quedado con los ultimos 100 aÃ±os(desde 1916 porque el dataset es solo hasta 2016)
''' 
df = df.drop(df[df['Year']<1916].index)
```

```python
df
```

```python
'''
Una vez que ya tengo una base de datos mas limpia y con los datos que a mi parecer eran los interesantes, 
he pensado en averiguar cuantos ataques ha habido por aÃ±o, en los ultimos 100 aÃ±os.
'''
df["Year"].value_counts().sort_index()
```

```python
'''
Sin embargo, para poder empezar a trabajar con la columna 'Country' tenia que cerciorarme de que no tuviera ningun valor nulo. 
Y claramente tenia demasiados. Para poderlo ver en perspectiva he creado un nuevo DataFrame que se llama 'data' en el que
tan solo me enseÃ±a las columnas relacionadas con el pais.
'''
data = df[df['Country'].isnull()]
data = data[['Country', 'Location', 'Area']]
data
```

```python
'''
Una vez tengo que filas son las que contienen valores nulos en 'Country', intentare hacer dos cosas:
1. Si la fila entera, es decir, 'Country', 'Area' y 'Location' esta vacia, la borrare por que sera imposible que yo pueda deducir 
donde exactamente han ocurrido esos ataques. Por lo tanto son filas que no me sirven para nada.
2. En el caso de que 'Area' o 'Location' contengan algun valor, podre deducir el pais en el que se encuentre dicha localizacion u area.
De tal manera que podre rellenar el valor nulo de la columna 'Country' con un valor de fiar.
'''
```

```python
'''
Opcion 1.
'''
nulos_everywhere = data[((data['Area'].isnull() & data['Location'].isnull())==True)]
nulos_everywhere
```

```python
'''
Como todas esas filas no hacen referencia a ningun lugar, ni dan una pista sobre donde se puede haber producido 
el/los ataques, pues las elimino.
'''

nulos = nulos_everywhere.index.tolist()
data_cl = df.drop(nulos, axis=0)
```

```python
#data_cl
'''
Comprobamos que de verdad ha eliminado todas aquellas filas que tenian nulos en 'Country', 'Area' y  'Location'.
Efectivamente, ha eliminado todas las filas correctamente.
'''
null_display = data_cl[(data_cl['Country'].isnull())]
null_display = null_display[['Country', 'Location', 'Area']]
null_display
```

```python
'''
Visto que son tan solo 11, lo cambio a mano buscando el location/area en google maps y asignandoles un pais.

1. Chekeamos que no haya otra fila en nuestro dataset que cumpla con las condiciones que nosotros queremos cambiar.
En caso de que haya mas de una, miraremos a ver que country han puesto ellos para esa misma location/area y pondremos el mismo.
En caso contrario, y que nuestra fila sea la unica que tiene esas caracteristicas en todo el data set, buscaremos en google a que pais pertence
la location/area determinada y lo rellenaremos.
En caso de que la informacion de la columna location/area no sea suficiente como para asignarle un pais/oceano, eliminare esas filas
por no contener datos que me vayan a servir en el futuro.
Este check lo he ido haciendo utilizando el siguiente codigo: ej. data_cl[data_cl['Country']=="SOUTH AMERICA"]

2. Checkeamos si de verdad hemos realizado el cambio correctamente: ej. data_cl.loc[3162, 'Country']
'''

data_cl.loc[(data_cl['Area']=='English Channel'), 'Country'] = 'ENGLAND'
data_cl.loc[(data_cl['Location']=='Between St. Kitts & Nevis'), 'Country'] = 'ST. KITTS & NEVIS' 
data_cl.loc[(data_cl['Location']=='Florida Strait'), 'Country'] = 'USA'
data_cl.loc[(data_cl['Area']=='Between Timor & Darwin, Australia'), 'Country'] = 'AUSTRALIA'
data_cl.loc[(data_cl['Area']=='Near the Andaman & Nicobar Islands'), 'Country'] = 'INDIA' #LO HE BUSCADO EN GOOGLE MAPS
data_cl.loc[(data_cl['Location']=='Geyser Bank'), 'Country'] = 'FRANCE'#ACTUALMENTE ESTA CONTROLADA POR FRANCIA PERO AMBOS COMOROS Y MADAGASCAR LO RECLAMAN COMO SUYO.
data_cl.loc[(data_cl['Location']=='ÃŽle Saint-Paul'), 'Country'] = 'FRANCE' #HE BUSCADO EN EL DATASET TODAS LAS SALIDAS CON EL COUNTRY==FRANCE Y HE VISTO QUE HAY OTRAS ISLAS COLONIA A LAS QUE SE LE HA ATRIBUIDO FRANCIA COMO PAIS.
data_cl.loc[(data_cl['Area']=='Caribbean Sea'), 'Country'] = 'CARIBBEAN SEA'
data_cl.loc[(data_cl['Location']=='225 miles east of Hong Kong'), 'Country'] = 'CHINA'
```

```python
data_cl.shape
```

```python
'''
Elimino las dos filas a las que no les puedo atribuir un pais concreto
'''
null_display = data_cl[(data_cl['Country'].isnull())]
null_display = null_display[['Country', 'Location', 'Area']]
null_display

nulos = null_display.index.tolist()
data_clean = data_cl.drop(nulos, axis=0)
```
