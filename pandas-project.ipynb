```python
# Voy a ver si puedo eliminar alguna columna de por si por contener muchos nulls.
print(df.shape)
null_cols = df.isnull().sum()
null_cols
```

```python
# Eliminar 'Unnamed:22' & 'Unnamed:23' porque casi todos sus valores son nulos.
drop_cols = list(null_cols[null_cols > 5000].index)
df = df.drop(drop_cols, axis=1)
print(df.shape)
```

```python
df.head()
```

```python
# Elimino tambien todas estas filas por que he pensado que no las quiero para nada.
df = df.drop(['pdf', 'href formula', 'href','Case Number.1','Case Number.2'], axis = 1)
```

```python
# Me quedo con los ultimos 100 años(desde 1916 porque el dataset es solo hasta 2016)
df = df.drop(df[df['Year']<1916].index)
```
